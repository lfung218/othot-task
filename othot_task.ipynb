## Data Transformations

__[Note]__  
Since we have no null values, we don't need to impute our dataset  
Also, we __chose not to standardized__ the data. The models we are going to build, such as Random Forest Classifier, are not distance-based. Even though we also have a logistic regression model with regularization, which usually requires feature scaling, we chose not to do feature scaling in order to be consistent and make it easier to interpret coefficients
### Create Dummies

```python
for col in cat_columns:
    dummies = pd.get_dummies(df[col], prefix=col, drop_first=True) # create n-1 dummy variables for each categorical variable
                                                                   # n is the number of values in the categorical variable
    df = pd.concat([df, dummies], axis=1)

leave_dummy = pd.get_dummies(df['LEAVE'], prefix='LEAVE')['LEAVE_LEAVE']
df = pd.concat([df, leave_dummy], axis=1)
```

```python
print('All columns in DataFrame: ')
print(df.columns.values)
```

## Data Partitioning

```python
input_columns = list(df.columns.values[1:8]) + list(df.columns.values[12:-1])
X = df[input_columns]
y = df.LEAVE_LEAVE

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
X_train.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
print('Trainning data size: {}'.format(X_train.shape[0]))
print('Testing data size: {}'.format(X_test.shape[0]))
```

## Model Building & Evaluation
Since we have a balanced dataset, __accuracy__ will be a decent scoring criteria for comparisons across different models. However, We will also look at metrics such as ROC AUC and f1 score and try to make our evalution more comprehensive and objective.

### Baseline Model
One possible baseline model for this task is a majority class classifier. This will simply classify all customers as stayers, since stayers are the majority. Let's apply a majority class classifer to our dataset and see what is the performance of it. Then, we will build a logistic regression model and a random forest classifier. Finally, we can compare these two models with our baseline model and select the one with the best performance

```python
from sklearn.model_selection import KFold
from sklearn.dummy import DummyClassifier
accuracy = 0

kf = KFold(n_splits=10, shuffle=True, random_state=0)
for train_indices, val_indices in kf.split(range(X_train.shape[0])): # use 10-fold cross validation to 
                                                                     # compute the average accuracy
    mc_clf = DummyClassifier()
    mc_clf.fit(X_train.iloc[train_indices], y_train.iloc[train_indices])
    pred_labels = mc_clf.predict(X_train.iloc[val_indices])
    accuracy += np.mean(pred_labels == y_train[val_indices])

print('Accuracy rate of the baseline model (Majority Class Classifier) is : {}'.format(accuracy/10))
```

#### Validation Result:
The baseline performance we want to beat is an accuarcy score of __0.504__

### Logistic Regression
Let's build logistic regression models to classify customers as potential "leavers" or "stayers". We will use 10-fold cross validation to tune hyperparameters. Then, we will use accuracy rate to determine which set of hyperparameters is the best. Finally, we will report the best average validation accuracy score based on the model we built using the best set of hyperparameters. We will also report the average ROC AUC and average f1 score for that model.

```python
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
from sklearn.metrics import f1_score
from scipy import interp

lr_highest_cross_val_accuracy = -np.inf
lr_hyperparams = [(C, penalty) for C in np.logspace(-3, 3, 7) for penalty in ['l1', 'l2']]

# sweep through the above hyperparameter settings. Find the best set in terms of accuracy
# also record the average ROC AUC and f-1 score under that best hyperparameter setting

for hyperparam in lr_hyperparams:
    candidate_C, candidate_penalty = hyperparam
    lr_accuracies = []
    lr_roc_aucs = []
    lr_avg_fpr = np.linspace(0, 1, 100)
    lr_tprs = []
    lr_f1s = []
    # perform 10 fold cross validation of current hyperparameter setting
    for train_indices, val_indices in kf.split(range(X_train.shape[0])):
        lr = LogisticRegression(penalty=candidate_penalty, C=candidate_C, solver='liblinear')
        lr.fit(X_train.iloc[train_indices], y_train.iloc[train_indices])
        lr_y_pred = lr.predict(X_train.iloc[val_indices])
        lr_y_pred_prob = lr.predict_proba(X_train.iloc[val_indices])[:, -1]
        lr_fpr, lr_tpr, _ = metrics.roc_curve(y_train.iloc[val_indices], lr_y_pred_prob)
        lr_tprs.append(interp(lr_avg_fpr, lr_fpr, lr_tpr)) # 
        lr_roc_auc = metrics.auc(lr_fpr, lr_tpr)
        accuracy = np.mean(lr_y_pred == y_train[val_indices])
        lr_f1 = f1_score(y_train.iloc[val_indices], lr_y_pred, average='weighted')  
        lr_accuracies.append(accuracy)
        lr_roc_aucs.append(lr_roc_auc)
        lr_f1s.append(lr_f1)
        
    cross_val_accuracy = np.mean(lr_accuracies)
    print('Hyperparam: ', hyperparam, ' Accuracy: ', cross_val_accuracy)
    
    # if the average accuracy score of this 10 fold cross validation is the highest
    if cross_val_accuracy > lr_highest_cross_val_accuracy:
        lr_highest_cross_val_accuracy = cross_val_accuracy
        lr_best_hyperparam = hyperparam
        lr_best_avg_auc = np.mean(lr_roc_aucs)
        lr_best_avg_tpr = np.mean(lr_tprs, axis=0)
        lr_best_f1 = np.mean(lr_f1s)
        lr_best_coef = lr.coef_ # record the coefficients of the logistic regression model we built using the 10th fold

print()
print('Best hyperparameter using accuracy as scoring method:', lr_best_hyperparam, 
      '\ncross validation accuracy:', lr_highest_cross_val_accuracy, '\naverage ROC AUC:', 
      lr_best_avg_auc, '\nf1 score: ', lr_best_f1)

# plot the ROC Curve for the best set of hyperparameter
plt.plot(lr_avg_fpr, lr_best_avg_tpr, label='ROC AUC=' + str(lr_best_avg_auc))
plt.plot([0, 1], [0, 1], linestyle='--', c='IndianRed')
plt.legend()
plt.title('ROC Curve when Accuracy is Highest')
plt.xlabel('FPR')
plt.ylabel('TPR')
```

Using 10-fold cross validation, we were able to find out the best set of hyperparameters. Let's try to take a look at the coefficients of the model we built in the last(10th) cross validation under the best hyperparameter setting and see if we can intepret these coefficients. However, since these coefficients are from a model which only learned 9/10 folds of the training data, they might not reflect the real-world situations

```python
lr_best_coef_pairs = {X.columns.values[i] : lr_best_coef.ravel()[i] for i in range(len(X.columns.values))}
lr_best_coef_pairs
```

#### Intepretation of Coefficients
+ The baseline for comparison will be a customer with 0 income, didn't use his/her quota at all, went to college, has an average satisfaction level, and actively looking for a new plan
+ Coefficients for Unsatisfied and Very Unsatisfied are relatively large. People who left are more likely to report they are unsatisfied
+ On the contrary, coefficient for satisified is less than zero and large in absolute value, meaning satisfied customers are less likely to leave
+ Coefficients for Over 15 Min Calls and Average Call Duration are also relatively large, meaning that customers who used more of their voice call quotas are more likely to leave
+ Coefficient for not attending college is negative and large in absolute value, meaning that customers who did not go to college are more likely to stay

#### Validation Results
The logistic regression model we chose has an accuracy rate of __0.638__, which is a __27% increase__ compared with the baseline model  
This model also has a ROC AUC of 0.692, and a f1 score of 0.637


### Random Forest Classifier
By applying a logistic regression model, we can improve our accuracy rate by 27%. Let's try to train an ensemble model, such as Random Forest, to our data and see if we can still improve on that. This time, we will also use accuracy as scoring criteria, since we will eventually use that to compare with our baseline model. However, we will still provide the average ROC AUC and average F1 score under the best set of parameters that yields the highest accuracy rate

```python
from sklearn.ensemble import RandomForestClassifier

rf_highest_cross_val_accuracy = -np.inf
rf_hyperparams = [(n_estimator, max_depth) for n_estimator in [50, 100, 150, 200] for max_depth in [3, 4, 5]]

# sweep through the above hyperparameter settings. Find the best set in terms of accuracy
# also record the average ROC AUC and f-1 score under that best hyperparameter setting
for hyperparam in rf_hyperparams:
    candidate_n_estimator, candidate_max_depth = hyperparam
    rf_accuracies = []
    rf_roc_aucs = []
    rf_avg_fpr = np.linspace(0, 1, 100)
    rf_tprs = []
    rf_f1s = []
    
    # perform 10 fold cross validation of current hyperparameter setting 
    for train_indices, val_indices in kf.split(range(X_train.shape[0])):
        rf = RandomForestClassifier(n_estimators=candidate_n_estimator, max_depth=candidate_max_depth)
        rf.fit(X_train.iloc[train_indices], y_train.iloc[train_indices])
        rf_y_pred = rf.predict(X_train.iloc[val_indices])
        rf_y_pred_prob = rf.predict_proba(X_train.iloc[val_indices])[:, -1]
        rf_fpr, rf_tpr, rf_threshold = metrics.roc_curve(y_train.iloc[val_indices], rf_y_pred_prob)
        rf_tprs.append(interp(rf_avg_fpr, rf_fpr, rf_tpr))
        rf_roc_auc = metrics.auc(rf_fpr, rf_tpr)
        rf_f1 = f1_score(y_train.iloc[val_indices], rf_y_pred, average='weighted')  
        accuracy = np.mean(rf_y_pred == y_train[val_indices])
        rf_accuracies.append(accuracy)
        rf_roc_aucs.append(rf_roc_auc)
        rf_f1s.append(rf_f1)

    cross_val_accuracy = np.mean(rf_accuracies)
    
    print('Hyperparam: ', hyperparam, ' Accuracy: ', cross_val_accuracy)
    
    # if the average accuracy score of this 10 fold cross validation is the highest
    if cross_val_accuracy > rf_highest_cross_val_accuracy:
        rf_highest_cross_val_accuracy = cross_val_accuracy
        rf_best_hyperparam = hyperparam
        rf_best_avg_auc = np.mean(rf_roc_aucs)
        rf_best_avg_tpr = np.mean(rf_tprs, axis=0)
        rf_best_f1 = np.mean(rf_f1s)
        rf_best_feat_imp = rf.feature_importances_ # record the feature importance of the random
                                                   # forest model we built using the 10th fold of data

print()
print('Best hyperparameter using accuracy as scoring method:', rf_best_hyperparam, '\ncross validation accuracy:', 
      rf_highest_cross_val_accuracy, '\naverage ROC AUC:', rf_best_avg_auc, '\nf1 score: ', rf_best_f1)

# plot the ROC curve for the best set of hyperparameter setting
plt.plot(rf_avg_fpr, rf_best_avg_tpr, label='ROC AUC=' + str(rf_best_avg_auc))
plt.plot([0, 1], [0, 1], linestyle='--', c='IndianRed')
plt.legend()
plt.title('ROC Curve when Accuracy is Highest')
plt.xlabel('FPR')
plt.ylabel('TPR')
```

After 10-fold cross validation for all possible combination of hyperparameters, we obtain the best average accuracy, as well as the ROC AUC and f1 score under that accuracy rate. We would like to check what features are more important. Thus, we retrieve the feature importances from the model we built in the last (10th) cross validation iteration. Below are the feature importances 

```python
feature_names = X_train.columns.values
feature_importance_pairs = {feature_names[i] : rf_best_feat_imp[i] for i in range(
                            rf_best_feat_imp.shape[0])}

# sort features by their importances
from operator import itemgetter

sorted_feature_importance_pairs = sorted(feature_importance_pairs.items(), reverse=True, key=itemgetter(1))
sorted_feature_importance_pairs
sorted_features = [feat for feat, imp in sorted_feature_importance_pairs]
sorted_importances = [imp for feat, imp in sorted_feature_importance_pairs]
```

```python
plt.bar(sorted_features, sorted_importances)
plt.xticks(rotation=90)
```
